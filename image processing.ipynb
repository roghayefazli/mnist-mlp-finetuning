```python
import numpy as np
import glob as gb
import cv2
```

```python
# path of every image in hoda folder 
image_path = gb.glob('Hoda 0-9/'+'*.bmp')
image_path
```

```python
x = []
y = []
for i in range(len(image_path)):
    path = image_path[i]
    idx = path.rfind('_')-1
    y.append(path[idx])
    img = cv2.imread(path,0)
    img = cv2.resize(img,dsize = (10,10))
    data = np.reshape(img,newshape=(100,))
    x.append(data)
x = np.array(x)
y = np.array(y)
```

```python
# shuffle data
list_per = np.random.permutation(len(x))
x_per = []
y_per = []
for i in range(len(list_per)):
    index_per = list_per[i]
    x_per.append(x[index_per])
    y_per.append(y[index_per])
x_per = np.array(x_per)
y_per = np.array(y_per)
```

```python
# train test split
train_test_index =int( 0.7*len(x_per))
x_train = x_per[0:train_test_index,:]
y_train = y_per[0:train_test_index]

x_test = x_per[train_test_index:,:]
y_test = y_per[train_test_index:]
```

```python
# normalize
x_train = x_train/255
x_test = x_test/255
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
mlp = MLPClassifier(hidden_layer_sizes=(50,20),activation='logistic',tol=0.0001,momentum=0.95,verbose=True,max_iter=500)
mlp.fit(x_train,y_train)
```

```python
print(mlp.score(x_train,y_train))
print(mlp.score(x_test,y_test))
```

```python
import joblib
joblib.dump(mlp,'mlp_network.joblib')
```

```python
new_mlp = joblib.load('mlp_network.joblib')
```

```python
print(new_mlp.score(x_train,y_train))
print(new_mlp.score(x_test,y_test))
```
